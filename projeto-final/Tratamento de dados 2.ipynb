```python
import pandas as pd
```

```python
pedalboards = pd.read_json('data/pedalboards.json')
pedalboards.columns = ['name', 0, 1, 2, 3, 4, 5]

pedalboards.head()
```

```python
plugins = pd.read_json('data/plugins.json')
plugins = plugins.sort_values(['key'])

plugins.head()
```

```python
#data = pd.DataFrame([], index=range(0, len(pedalboards)), columns=plugins.name)
data = pd.DataFrame([], index=range(0, len(pedalboards)), columns=plugins.key)
data = data.fillna(0)

data
```

```python
for index, pedalboard in pedalboards[[0, 1, 2, 3, 4, 5]].iterrows():
    for effect in pedalboard:
        data[effect][index] += 1

data.head()    
```
