```python
import pandas as pd
pedalboards = pd.read_json('data/pedalboards-list.json', orient='columns')

pedalboards.head()
```

```python
len(pedalboards)
```

```python
import requests
import xmltodict

def read_pedalboard(index):
    response = requests.get('http://guitarpatches.com/download.php?unit=G3&mode=download&ID={}'.format(index))
    return xmltodict.parse(response.content)

pedalboard = read_pedalboard(pedalboards.iloc[0]['index'])
pedalboard
```

```python
def extract_data(pedalboard):
    data = [pedalboard['PatchData']['Name']]

    for index in range(0, 6):
        data.append(pedalboard['PatchData']['Module{}'.format(index)]['Prm1'])
    
    return data

extract_data(pedalboard)
```

```python
data = []
errors = []

i = 0
for index in pedalboards['index']:
    print(i, index)
    i += 1

    try:
        pedalboard = read_pedalboard(index)
        data.append(extract_data(pedalboard))
    except:
        errors.append(index)
```
